This notebook is for gathering basic information about the datasets FNDatasetGerman and GermanFakeNC, 
such as categories, unique URLs, how many articles of fake and legitimate news, etc. 

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import numpy as np
import pandas as pd

from pprint import pprint
```

Analysis of the FNDatasetGerman dataset

```python
FNDatasetGerman = pd.read_csv(r"datasets/FNDatasetGer.csv",encoding="utf-8")
FNDatasetGerman
```

```python
FNDatasetGerman.nunique(dropna=False)
```

```python
uniques = dict()
for colname in ["Kategorie","Quelle","Fake","Art"]:
    uniques[colname] = FNDatasetGerman[colname].unique()
```

```python
uniques
```

```python
grouping = FNDatasetGerman.groupby("Fake")
realNewsFNDG = grouping.get_group(0)
fakeNewsFNDG = grouping.get_group(1)
```

```python
len(realNewsFNDG)
```

```python
suniques = dict()
for news,fake in zip([realNewsFNDG,fakeNewsFNDG],("Real","Fake")):
    for colname in ["Kategorie","Quelle","Art"]:
        suniques[(fake,colname)] = news[colname].value_counts(dropna=False)
```

```python
pprint(suniques)
```

Now for the analysis of GermanFakeNC:


```python
germanFakeNC = pd.read_json(r"datasets/GermanFakeNC.json",encoding="utf-8")
germanFakeNC
```
